### 1) curve plotting

```python
#curve plotting part (a)
# MPL
import matplotlib        as mpl
import matplotlib.pyplot as plt

# NumPy
import numpy as np
from numpy import cos, exp, pi, sin, sinc, sinh, sqrt

mpl.rcParams['legend.frameon'] = False

#parametric equation x(theta) and y(theta) given 0 to 2pi theta
t=np.linspace(0,2*pi,100)
x=2*cos(t)+cos(2*t)
y=2*sin(t)-sin(2*t)

plt.figure()
plt.plot(x, y, 'b-')


plt.xlabel('x = 2cos(t) + cos(2t)')
plt.ylabel('y = 2sin(t) - sin(2t)')
plt.title('x(t) vs y(t) over t from 0 to 2Pi inclusive')
#plt.legend(loc=0)

plt.show()
```

```python
#curve plotting part (b)
# MPL
import matplotlib        as mpl
import matplotlib.pyplot as plt

# NumPy
import numpy as np
from numpy import cos, exp, pi, sin, sinc, sinh, sqrt

mpl.rcParams['legend.frameon'] = False

#make a polar plot r=f(t) and then make a cartesian plot by converting r 
#to x and y coordinates with x=rcos(t) y=rsin(t)
#use this method for r=t^2 for t:[0,10pi]

t=np.linspace(0,10*pi,300)
r=t**2
x=r*cos(t)
y=r*sin(t)

plt.figure()
plt.plot(x, y, 'b-')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Galilean spiral r = t^2 for t : [0, 10Pi]')


plt.show()

```

```python
#curve plotting part (c)
# MPL
import matplotlib        as mpl
import matplotlib.pyplot as plt

# NumPy
import numpy as np
from numpy import cos, exp, pi, sin, sinc, sinh, sqrt

mpl.rcParams['legend.frameon'] = False

t=np.linspace(0,24*pi,2000)
r=exp(cos(t))-2*cos(4*t)+sin(t/12)**5
x=r*cos(t)
y=r*sin(t)

plt.figure()
plt.plot(x, y, 'b-')


plt.xlabel('x')
plt.ylabel('y')
plt.title("Fey's function r = e^cos(t) - 2cos(4t) + sin(t/12)^5 for t : [0, 24Pi]")


plt.show()
```

### 2) Mandlebrot Set
